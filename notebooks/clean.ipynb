```python
#Importar bibliotecas
import pandas as pd 
import numpy as np
```

```python
#Ler o CSV
df = pd.read_csv('../data/nhl_draft_predraft_stats.csv', sep=';')
df.head()
```

```python
#Verificar existência de caracteres especiais
mask = df['player'].str.contains(r'[^\x00-\x7F]', na=False) | df['team'].str.contains(r'[^\x00-\x7F]', na=False)
df[mask][['player', 'team']]
```

```python
#Exibir os valores na coluna "position"
df['position'].value_counts()
```

```python
#Exibir o número de linhas e colunas do CSV
df.shape
```

```python
#Remover todas as linhas que correspondiam ao G (goleiro)
df = df[df['position'] != 'G']
```

```python
#Exibir os valores na coluna "position" após a limpeza
df['position'].value_counts()
```

```python
#Exibir o número de linhas e colunas do CSV após a limpeza
df.shape
```

```python
#Exibir quantas colunas apresentavam valores "vazios"
print(df.isnull().sum())
```

```python
#Remover as linhas que continham valores "vazios" nessas colunas
df = df.dropna(subset=['amateur_league', 'seasons_pre_nhl', 'games_played', 'goals', 'assists', 'points', 'penalty_minutes', 'points_per_game', 'goals_per_game'])
```

```python
#Exibir quantas colunas apresentavam valores "vazios" após a limpeza
print(df.isnull().sum())
```

```python
#Salvar os dados limpos e um novo CSV
df.to_csv('../data/nhl_draft_clean.csv', index=False)
```
